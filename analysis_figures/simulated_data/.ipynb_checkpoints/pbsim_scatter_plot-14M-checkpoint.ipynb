```python
import numpy as np
import pandas as pd
import pysam
import multiprocessing
import matplotlib.pyplot as plt
from pydeseq2.preprocessing import deseq2_norm
import seaborn as sns
import csv
from os.path import join
from scipy.stats import pearsonr
from sklearn.metrics import precision_score, recall_score
import json
from sklearn.metrics import roc_curve, auc
from sklearn.metrics import precision_recall_curve, PrecisionRecallDisplay, auc, average_precision_score
from scipy.io import mmread
```

```python
def spearman_count(quant_dir):
    
    
    d = ["SQ2", "RSII"]
    ground_truth = {"SQ2": {}, "RSII":{}}
    t = {"SQ2": {}, "RSII":{}}
    
    
    #get the ground truth values
    txp = pd.read_csv("/mnt/scratch2/zahra/sim_quant/badread_quant/ground_truth/all_transcript_name.csv", header=None, sep="\t")
    first_parts = [s.split('.')[0] for s in list(txp[0])]
    first_parts_df = pd.DataFrame(first_parts)
    first_parts_df.columns = ["Txp"]
    
    
    for k in d:
        ground_truth[k] = pd.read_csv(f"/mnt/scratch2/zahra/pbsim_quant/ground_truth/{k}_file.csv", header=None, sep="\t")
        ground_truth[k].columns = ["Txp", "forward", "reverse"]
        ground_truth[k]["Ground_Truth"] = ground_truth[k]["forward"] + ground_truth[k]["reverse"]
        ground_truth[k]["Txp"] = [s.split('.')[0] for s in list(ground_truth[k]["Txp"])]
        
    
    for k in d:
        t[k] = pd.merge(first_parts_df, ground_truth[k], left_on="Txp", right_on="Txp", how="left")
        t[k] = t[k].fillna(0) 
    
    
    
    #read in the count files   
    oarfish_v5_cov = {k: {} for k in d}
    oarfish_v5_nocov = {k: {} for k in d}
    NanoCount = {k: {} for k in d}
    NanoCount_nofilt = {k: {} for k in d}
    bambu = {k: {} for k in d}
    kallisto = {k: {} for k in d}
    isoquant = {k: {} for k in d}
    transigner = {k: {} for k in d}
    
    
    for k in d:
        oarfish_v5_cov[k] = pd.read_csv(f"{quant_dir}/oarfish_quant/{k}/out_cov.quant", sep="\t")
        oarfish_v5_cov[k]['tname'] = oarfish_v5_cov[k]['tname'].apply(lambda x: x.split('.')[0])
        ###########################################################################################
        oarfish_v5_nocov[k] = pd.read_csv(f"{quant_dir}/oarfish_quant/{k}/out_nocov.quant", sep="\t")
        oarfish_v5_nocov[k]['tname'] = oarfish_v5_nocov[k]['tname'].apply(lambda x: x.split('.')[0])
        ##########################################################################################
        NanoCount[k] = pd.read_csv(f"{quant_dir}/NanoCount_quant/{k}/isoform_expression_estimates", sep="\t")
        NanoCount[k].rename(columns={NanoCount[k].columns[3]: "num_reads"}, inplace=True)
        NanoCount[k]['transcript_name'] = NanoCount[k]['transcript_name'].apply(lambda x: x.split('.')[0])
        NanoCount[k] = NanoCount[k].groupby(NanoCount[k]['transcript_name'])['num_reads'].sum().reset_index()
        ###########################################################################################
        NanoCount_nofilt[k] = pd.read_csv(f"{quant_dir}/NanoCount_quant_nofilt/{k}/isoform_expression_estimates", sep="\t")
        NanoCount_nofilt[k].rename(columns={NanoCount_nofilt[k].columns[3]: "num_reads"}, inplace=True)
        NanoCount_nofilt[k]['transcript_name'] = NanoCount_nofilt[k]['transcript_name'].apply(lambda x: x.split('.')[0])
        NanoCount_nofilt[k] = NanoCount_nofilt[k].groupby(NanoCount_nofilt[k]['transcript_name'])['num_reads'].sum().reset_index()
        ###########################################################################################
        bambu[k] = pd.read_csv(f"{quant_dir}/bambu_quant/{k}/counts_transcript.txt", sep="\t")
        bambu[k].rename(columns={bambu[k].columns[2]: "num_reads"}, inplace=True)
        bambu[k]['TXNAME'] = bambu[k]['TXNAME'].apply(lambda x: x.split('.')[0])
        bambu[k] = bambu[k].groupby(bambu[k]['TXNAME'])['num_reads'].sum().reset_index()
        ###########################################################################################
        count = mmread(f'{quant_dir}/lr-kallisto/{k}/matrix.abundance.mtx')
        labels = pd.read_csv(f'{quant_dir}/lr-kallisto/{k}/transcripts.txt', header=None, sep='\t')
        kallisto[k] = pd.DataFrame(count.todense().T, columns=['lr-Kallisto'])
        kallisto[k]['transcript_id'] = [labels.values[i][0] for i in range(np.shape(labels.values)[0])]
        kallisto[k]['transcript_id'] = kallisto[k]['transcript_id'].str.split('.').str[0]
        kallisto[k].rename(columns={kallisto[k].columns[0]: "num_reads"}, inplace=True)
        kallisto[k] = kallisto[k].groupby(kallisto[k]['transcript_id'])['num_reads'].sum().reset_index()
        ###########################################################################################
        transigner[k] = pd.read_csv(f"{quant_dir}/transigner_quant/{k}/abundances.tsv", sep="\t", header=None)
        transigner[k].columns = ['Txp', 'abundance', 'num_reads']
        transigner[k].rename(columns={transigner[k].columns[2]: "num_reads"}, inplace=True)
        transigner[k]['Txp'] = transigner[k]['Txp'].apply(lambda x: x.split('.')[0])
        transigner[k] = transigner[k].groupby(transigner[k]['Txp'])['num_reads'].sum().reset_index()
        ###########################################################################################
        isoquant[k] = pd.read_csv(f"{quant_dir}/IsoQuant_quant/{k}/OUT/OUT.transcript_tpm.tsv", sep="\t")
        isoquant[k]['#feature_id'] = isoquant[k]['#feature_id'].str.split('.').str[0]
        isoquant[k].rename(columns={isoquant[k].columns[1]: "num_reads"}, inplace=True)
        isoquant[k] = isoquant[k].groupby(isoquant[k]['#feature_id'])['num_reads'].sum().reset_index()
        


    m = {k: {} for k in d}
    
    for v in d:
        print(v)
        m[v] = pd.merge(oarfish_v5_cov[v], t[v], left_on="tname", right_on="Txp", how="left", suffixes=('_v5_cov', ''))
        m[v] = m[v].rename(columns={"num_reads" : "oarfish 0.5 (cov)"})
        print(len(m[v]["tname"]))
        #--------------------------------------------------------------------------------------------------
        m[v] = pd.merge(oarfish_v5_nocov[v], m[v], left_on="tname", right_on="tname", how="right", suffixes=('_v5_nocov', ''))
        m[v] = m[v].rename(columns={"num_reads" : "oarfish 0.5 (nocov)"})
        print(len(m[v]["tname"]))
        #--------------------------------------------------------------------------------------------------
        m[v] = pd.merge(NanoCount[v], m[v], left_on="transcript_name", right_on="tname", how="right", suffixes=('_NanoCount', ''))
        m[v] = m[v].rename(columns={"num_reads" : "NanoCount"})
        print(len(m[v]["tname"]))
        #--------------------------------------------------------------------------------------------------
        m[v] = pd.merge(NanoCount_nofilt[v], m[v], left_on="transcript_name", right_on="tname", how="right", suffixes=('_NanoCount_nofilt', ''))
        m[v] = m[v].rename(columns={"num_reads" : "NanoCount (nofilt)"})
        print(len(m[v]["tname"]))
        #--------------------------------------------------------------------------------------------------
        m[v] = pd.merge(bambu[v], m[v], left_on="TXNAME", right_on="tname", how="right", suffixes=('_bambu', ''))
        m[v] = m[v].rename(columns={"num_reads" : "bambu"})
        print(len(m[v]["tname"]))
        #--------------------------------------------------------------------------------------------------
        m[v] = pd.merge(kallisto[v], m[v], left_on="transcript_id", right_on="tname", how="right", suffixes=('_kallisto', ''))
        m[v] = m[v].rename(columns={"num_reads" : "lr-kallisto"})
        print(len(m[v]["tname"]))
        #--------------------------------------------------------------------------------------------------
        m[v] = pd.merge(transigner[v], m[v], left_on="Txp", right_on="tname", how="right", suffixes=('_transigner', ''))
        m[v] = m[v].rename(columns={"num_reads" : "TranSigner"})
        print(len(m[v]["tname"]))
        #--------------------------------------------------------------------------------------------------
        m[v] = pd.merge(isoquant[v], m[v], left_on="#feature_id", right_on="tname", how="right", suffixes=('_isoquant', ''))
        m[v] = m[v].rename(columns={"num_reads" : "IsoQuant"})
        print(len(m[v]["tname"]))
        #--------------------------------------------------------------------------------------------------
        m[v] = m[v].fillna(0)
    

    columns_to_compare = ["oarfish 0.5 (cov)", "oarfish 0.5 (nocov)", "NanoCount", "NanoCount (nofilt)", "bambu", "lr-kallisto", "TranSigner", "IsoQuant"]
    epsilon = 1e-10
    
    return m
    
```

```python
quant_dir = "/mnt/scratch2/zahra/pbsim_quant"

#obtain the dataframe of quants for different categories
Quant = spearman_count(quant_dir)
```

```python
#infrv = 0.1 (Hct116 vs. short read)

method_names = ["oarfish 0.5 (cov)", "oarfish 0.5 (nocov)", "NanoCount", "NanoCount (nofilt)", "bambu", "lr-kallisto", "TranSigner", "IsoQuant"]

small_constant = 1e-10
# Create a 2x4 grid of subplots
fig, axs = plt.subplots(2, 4, figsize=(15, 10))

for i, method_name in enumerate(method_names):
    # Determine subplot position based on the row and column
    row = i // 4
    col = i % 4
    
    illumina_value = np.log2(Quant["SQ2"]["Ground_Truth"] + 1.0)
    ont_value = np.log2(Quant["SQ2"][method_name] + 1.0)

    # Create a hexbin plot
    hb = axs[row, col].hexbin(illumina_value, ont_value, gridsize=(20, 20), cmap='turbo', bins='log', mincnt=1)

    # Calculate Pearson correlation coefficient and p-value
    r_value, p_value = pearsonr(illumina_value, ont_value)

    # Fit a linear regression line
    slope, intercept = np.polyfit(illumina_value, ont_value, 1)

    # Add the regression line
    x_values = np.linspace(np.min(illumina_value), np.max(illumina_value), 100)
    y_values = slope * x_values + intercept
    axs[row, col].plot(x_values, y_values, color='red', label=f'Regression Line\nPearson R: {r_value:.2f}')

    # Add method name bar at the top of each plot
    axs[row, col].text(0.5, 1.03, method_name, ha='center', va='center', transform=axs[row, col].transAxes, bbox=dict(facecolor='white', edgecolor='none', boxstyle='round,pad=0.3', alpha=0.5), fontsize=12)
    
    # Display the legend
    axs[row, col].legend(fontsize=11)

# Add common colorbar
cb_ax = fig.add_axes([0.95, 0.15, 0.02, 0.7])  # [left, bottom, width, height]
cb = fig.colorbar(hb, cax=cb_ax, cmap="turbo")
cb.set_label('log\u2082(Read Count)', fontsize=14)

# Set common x and y labels
fig.text(0.5, 0.04, 'log\u2082(Ground-Truth counts + 1.0)', ha='center', va='center', fontsize=16)
fig.text(0.06, 0.5, 'log\u2082(ts-cDNA Long-Read counts + 1.0)', ha='center', va='center', rotation='vertical', fontsize=16)

# Show the plot
plt.show()
```

```python

method_names = ["oarfish 0.5 (cov)", "oarfish 0.5 (nocov)", "NanoCount", "NanoCount (nofilt)", "bambu", "lr-kallisto", "TranSigner", "IsoQuant"]

small_constant = 1e-10
# Create a 2x3 grid of subplots
fig, axs = plt.subplots(2, 4, figsize=(15, 10))

for i, method_name in enumerate(method_names):
    # Determine subplot position based on the row and column
    row = i // 4
    col = i % 4
    
    # Ground truth (binary labels)
    true = [1 if x > 0 else 0 for x in Quant["SQ2"]["Ground_Truth"].values]
    
    # Prediction scores
    pred = Quant["SQ2"][method_name].values

    # Create precision-recall display
    precision, recall, _ = precision_recall_curve(true, pred)
    
    # Calculate AUC for precision-recall curve
    pr_auc = auc(recall, precision)
    
    # Calculate and display average precision score
    avg_precision = average_precision_score(true, pred)
    
    # Plot precision-recall curve using PrecisionRecallDisplay plot method
    disp = PrecisionRecallDisplay(precision=precision, recall=recall)
    disp.plot(ax=axs[row, col])
    
    # Add method name bar at the top of each plot
    axs[row, col].text(0.5, 1.05, method_name, ha='center', va='center', transform=axs[row, col].transAxes, bbox=dict(facecolor='white', edgecolor='none', boxstyle='round,pad=0.3', alpha=0.5), fontsize=12)
    
    # Add AUC to the plot
    # Add legend with AUC and avg_precision
    legend_text = f'AUC: {pr_auc:.2f}\nAvg Precision: {avg_precision:.2f}'
    axs[row, col].text(0.95, 0.95, legend_text, ha='right', va='top', transform=axs[row, col].transAxes, fontsize=11, bbox=dict(facecolor='none', edgecolor='black', boxstyle='round,pad=0.3'))
    

    #axs[row, col].text(0.5, 0.1, f'Avg Precision: {avg_precision:.2f}', ha='center', va='center', transform=axs[row, col].transAxes)
    

# Set common x and y labels
fig.text(0.5, -0.05, 'Recall', ha='center', va='center', fontsize=16)
fig.text(0.0, 0.5, 'Precision', ha='center', va='center', rotation='vertical', fontsize=16)

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 1])

# Show the plot
plt.show()
```
