```python
import os
import re
import matplotlib.pyplot as plt
import pandas as pd
from os.path import join
import seaborn as sns
from matplotlib.gridspec import GridSpec
```

```python
def extract_max_resident_set_size(file_path):
    with open(file_path, 'r') as file:
        content = file.read()
        match = re.search(r"Maximum resident set size \(kbytes\): (\d+)", content)
        if match:
            return int(match.group(1))
        else:
            return None
```

```python
def extract_wall_clock_time(file_path):
    with open(file_path, 'r') as file:
        content = file.read()
        match = re.search(r'Elapsed \(wall clock\) time \(h:mm:ss or m:ss\): (\S+)', content)
        if match:
            wall_clock_time = match.group(1)
            #print(wall_clock_time)
            # Split the time into hours, minutes, and seconds
            time_parts = wall_clock_time.split(':')
            
            # Initialize variables for hours, minutes, and seconds
            hours = 0
            minutes = 0
            seconds = 0

            # Assign values based on the number of time parts
            if len(time_parts) == 3:
                hours = int(time_parts[0])
                minutes = int(time_parts[1])
                seconds = float(time_parts[2])
            elif len(time_parts) == 2:
                minutes = int(time_parts[0])
                seconds = float(time_parts[1])
            elif len(time_parts) == 1:
                seconds = float(time_parts[0])

            # Convert to total seconds
            total_seconds = hours * 3600 + minutes * 60 + seconds
            return total_seconds
        else:
            return None
```

```python
def plot_function(input_list, fig_path):
    
    fig = plt.figure(figsize=(12,10))
    gs = GridSpec(2, 3, width_ratios=[1, 1, 0.2])  # 2 rows, 3 columns; last column for legend

    ax1 = fig.add_subplot(gs[0, 0])
    ax2 = fig.add_subplot(gs[0, 1])
    ax3 = fig.add_subplot(gs[1, 0])
    ax4 = fig.add_subplot(gs[1, 1])
    ax_legend = fig.add_subplot(gs[:, 2])  # This will be the space for the legend
    ax_legend.axis('off')  # Turn off the axis for the legend
    
    color_palette = sns.color_palette("dark", 5)
    marker_palette = ['o', '^', 's', 'p', 'h', '+', 'x', '*']
    
    legend_handles = []
    legend_labels = []
    
    cl = 0
    ml = 0
    
    for v1, v2 in {"H9_directRNA": "direct RNA (H9)", 'SQ2': 'SQ2 HiFi'}.items():
        line, = ax1.plot(list(input_list[v1]['Method']), list(input_list[v1]['time']), 
                         marker=marker_palette[ml], ms=6, linestyle='-', color=color_palette[cl], label=f"{v2}")
        legend_handles.append(line)
        legend_labels.append(f"{v2}")
        cl += 1
        ml += 1
    ax1.set_xlabel('Method', fontsize=10)
    ax1.set_ylabel('Elapsed Time (seconds)', fontsize=10)
    ax1.set_title('Elapsed Time for all methods', fontsize=10)
    ax1.grid(True)  # Add grid lines
    ax1.set_xticklabels(list(input_list[v1]['Method']), rotation=45, fontsize=10, ha='right')  # Rotate x-axis labels for readability
    ax1.text(0.0, 1.05, 'a)', transform=ax1.transAxes, ha="left", va="center", fontsize=12, weight="bold")
    
    cl = 0
    ml = 0
    
    for v1, v2 in {"H9_directRNA": "direct RNA (H9)", 'SQ2': 'SQ2 HiFi'}.items():
        line, = ax2.plot(list(input_list[v1]['Method']), list(input_list[v1]['memory']), 
                         marker=marker_palette[ml], ms=6, linestyle='-', color=color_palette[cl], label=f"{v2}")
        cl += 1
        ml += 1
    ax2.set_xlabel('Method', fontsize=10)
    ax2.set_ylabel('Memory Usage (kbytes)', fontsize=10)
    ax2.set_title('Memory Usage for all methods', fontsize=10)
    ax2.grid(True)  # Add grid lines
    ax2.set_xticklabels(list(input_list[v1]['Method']), rotation=45, fontsize=10, ha='right')  # Rotate x-axis labels for readability
    ax2.text(0.0, 1.1, 'b)', transform=ax2.transAxes, ha="left", va="center", fontsize=12, weight="bold")
    
    cl = 0
    ml = 0
    
    for v1, v2 in {"H9_directRNA": "direct RNA (H9)", 'SQ2': 'SQ2 HiFi'}.items():
        line, = ax3.plot(list(input_list[v1]['Method'][:-1]), list(input_list[v1]['time'][:-1]), 
                         marker=marker_palette[ml], ms=6, linestyle='-', color=color_palette[cl], label=f"{v2}")
        cl += 1
        ml += 1
    ax3.set_xlabel('Method', fontsize=10)
    ax3.set_ylabel('Elapsed Time (seconds)', fontsize=10)
    ax3.set_title('Elapsed Time (Excludes ESPRESSO)', fontsize=10)
    ax3.grid(True)  # Add grid lines
    ax3.set_xticklabels(list(input_list[v1]['Method'][:-1]), rotation=45, fontsize=10, ha='right')  # Rotate x-axis labels for readability
    ax3.text(0.0, 1.05, 'c)', transform=ax3.transAxes, ha="left", va="center", fontsize=12, weight="bold")
    
    cl = 0
    ml = 0
    
    for v1, v2 in {"H9_directRNA": "direct RNA (H9)", 'SQ2': 'SQ2 HiFi'}.items():
        line, = ax4.plot(list(input_list[v1]['Method'][:-1]), list(input_list[v1]['memory'][:-1]), 
                         marker=marker_palette[ml], ms=6, linestyle='-', color=color_palette[cl], label=f"{v2}")
        cl += 1
        ml += 1
    ax4.set_xlabel('Method', fontsize=10)
    ax4.set_ylabel('Memory Usage (kbytes)', fontsize=10)
    ax4.set_title('Memory Usage (Excludes ESPRESSO)', fontsize=10)
    ax4.grid(True)  # Add grid lines
    ax4.set_xticklabels(list(input_list[v1]['Method'][:-1]), rotation=45, fontsize=10, ha='right')  # Rotate x-axis labels for readability
    ax4.text(0.0, 1.1, 'd)', transform=ax4.transAxes, ha="left", va="center", fontsize=12, weight="bold")
    
    plt.tight_layout()
    
    # Create the single legend outside the subplots
    ax_legend.legend(handles=legend_handles, labels=legend_labels, loc='center', fontsize=10)

    # Save the figure as a JPEG file
    directory = os.path.dirname(fig_path)
    if not os.path.exists(directory):
        os.makedirs(directory)
    plt.savefig(fig_path, format='jpg', dpi=300)
    
    plt.show()
```

```python
quant_ont = '/fs/cbcb-scratch/zzare/sim_quant/nanosim_quant'
quant_pacbio = '/fs/cbcb-scratch/zzare/sim_quant/tksm_quant_new_model'

data_dict = {
    "H9_directRNA": ("drna", quant_ont),
    "SQ2": ("", quant_pacbio)
    }

method_list = ["oarfish align (cov)", "oarfish align (nocov)", "oarfish raw (cov)", "oarfish raw (nocov)", "NanoCount", "NanoCount (nofilt)", "bambu", "lr-kallisto", "TranSigner", "IsoQuant", "ESPRESSO"]

final_df = {}

for k, (v, quant_dir) in data_dict.items():
    final_df[k] = pd.DataFrame()
    final_df[k]["Method"] = method_list
    
    #shr dataset
    file_path = {}
    file_path = {'oarfish align (cov)':     join(quant_dir, 'oarfish_align_quant', k, v, 'time_cov'),
                 'oarfish align (nocov)':   join(quant_dir, 'oarfish_align_quant', k, v, 'time_nocov'),
                 'oarfish raw (cov)':       join(quant_dir, 'oarfish_raw_quant', k, v, 'time_cov'),
                 'oarfish raw (nocov)':     join(quant_dir, 'oarfish_raw_quant', k, v, 'time_nocov'),
                 'NanoCount':               join(quant_dir, 'NanoCount_quant', k, v, 'time_info'),
                 'NanoCount (nofilt)':      join(quant_dir, 'NanoCount_quant_nofilt', k, v, 'time_info'),
                 'bambu':                   join(quant_dir, 'bambu_quant', k, v, 'time_info'),
                 'lr-kallisto':             join(quant_dir, 'lr-kallisto', k, v),
                 'TranSigner':              join(quant_dir, 'transigner_quant', k, v),
                 'IsoQuant':                join(quant_dir, 'IsoQuant_quant', k, v, 'time_info'),
                 'ESPRESSO':                join(quant_dir, 'espresso_quant', k, v)}
    if v != "":
        align_path = {'align_T':                 join(quant_dir, f"alignment/{k}/{v}_time_T"),
                      'align_G':                 join(quant_dir, f"alignment/{k}/{v}_time_G")}
    else:
        align_path = {'align_T':                 join(quant_dir, f"alignment/{k}_time_T"),
                      'align_G':                 join(quant_dir, f"alignment/{k}_time_G")}

    if k == "SQ2":
        idx_path = {'idx_T':                 join(quant_dir, f"alignment/{k}_idx_time_T"),
                    'idx_G':                 join(quant_dir, f"alignment/idx_time_G")}
    else:
        idx_path = {'idx_T':                 join(quant_dir, f"alignment/idx_time_T"),
                    'idx_G':                 join(quant_dir, f"alignment/idx_time_G")}
    
    elapsed_time = []
    memory_usage = []
    for method, path in file_path.items():
        if method in ["oarfish align (cov)", "oarfish align (nocov)"]:
            elapsed_time.append(extract_wall_clock_time(path) + extract_wall_clock_time(align_path['align_T']))
            memory_usage.append(extract_max_resident_set_size(path) + extract_max_resident_set_size(align_path['align_T']))
            
        elif method in ["oarfish raw (cov)", "oarfish raw (nocov)"]:
            elapsed_time.append(extract_wall_clock_time(path) - extract_wall_clock_time(idx_path['idx_T']))
            memory_usage.append(extract_max_resident_set_size(path) - extract_max_resident_set_size(idx_path['idx_T']))
            
        elif method in ["NanoCount", "NanoCount (nofilt)"]:
            elapsed_time.append(extract_wall_clock_time(path) + extract_wall_clock_time(align_path['align_T']))
            memory_usage.append(extract_max_resident_set_size(path) + extract_max_resident_set_size(align_path['align_T']))

        elif method == "bambu":
            elapsed_time.append(extract_wall_clock_time(path) + extract_wall_clock_time(align_path['align_G']))
            memory_usage.append(extract_max_resident_set_size(path) + extract_max_resident_set_size(align_path['align_G']))

        elif method == "lr-kallisto":
            elapsed_time.append(extract_wall_clock_time(path + '/time_kallisto_bus') + extract_wall_clock_time(path + '/time_bustools_sort') + extract_wall_clock_time(path + '/time_bustools_count') + extract_wall_clock_time(path + '/time_kallisto_quant'))
            memory_usage.append(extract_max_resident_set_size(path + '/time_kallisto_bus') + extract_max_resident_set_size(path + '/time_bustools_sort') + extract_max_resident_set_size(path + '/time_bustools_count') + extract_max_resident_set_size(path + '/time_kallisto_quant'))
            
        elif method == "TranSigner":
            elapsed_time.append(extract_wall_clock_time(path+"/time_align") + extract_wall_clock_time(path+"/time_prefilter") + extract_wall_clock_time(path+"/time_em") - extract_wall_clock_time(idx_path['idx_T']))
            memory_usage.append(extract_max_resident_set_size(path+"/time_align") + extract_max_resident_set_size(path+"/time_prefilter") + extract_max_resident_set_size(path+"/time_em") - extract_max_resident_set_size(idx_path['idx_T'])) 

        elif method == "IsoQuant":
            elapsed_time.append(extract_wall_clock_time(path) - extract_wall_clock_time(idx_path['idx_G']))
            memory_usage.append(extract_max_resident_set_size(path) - extract_max_resident_set_size(idx_path['idx_G']))
            
        elif method == "ESPRESSO":
            elapsed_time.append(extract_wall_clock_time(path+"/time_ESPRESSO_S") + extract_wall_clock_time(path+"/time_ESPRESSO_C") + extract_wall_clock_time(path+"/time_ESPRESSO_Q") + extract_wall_clock_time(path+"/time_sort") + extract_wall_clock_time(align_path['align_G']))
            memory_usage.append(extract_max_resident_set_size(path+"/time_ESPRESSO_S") + extract_max_resident_set_size(path+"/time_ESPRESSO_C") + extract_max_resident_set_size(path+"/time_ESPRESSO_Q") + + extract_max_resident_set_size(path+"/time_sort") + extract_max_resident_set_size(align_path['align_G']))            
        
            
    final_df[k]["time"] = elapsed_time
    final_df[k]["memory"] = memory_usage
    
```
