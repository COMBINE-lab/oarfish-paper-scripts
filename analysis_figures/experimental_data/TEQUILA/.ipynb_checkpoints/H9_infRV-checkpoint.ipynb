```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read in bootstrap values of benchmarking dataset
```

```python
boot_H9 = pd.read_csv("/mnt/scratch2/zahra/exp_quant_new/H9_quant/short_read/H9_21/aux_info/bootstrap/quant_bootstraps.tsv", sep="\t")
```

```python
boot_Hct116 = boot_Hct116.T
```

```python
boot_Hct116
```

```python
boot_Hct116.index
```

```python
row_variances = boot_Hct116.var(axis=1)
row_variances
```

```python
row_means = boot_Hct116.mean(axis=1)
row_means
```

```python
row_variances = np.array(row_variances)
row_means = np.array(row_means)

# Calculate InfRV with a condition to handle row_means equal to zero
infRV_Hct116 = np.zeros_like(row_means)  # Initialize with zeros

non_zero_mask = row_means != 0
infRV_Hct116[non_zero_mask] = np.maximum(row_variances[non_zero_mask] - row_means[non_zero_mask], 0) / row_means[non_zero_mask]

print(infRV_Hct116)
```

```python
infRV_Hct116_df = {}
infRV_Hct116_df["Txp"] = boot_Hct116.index
infRV_Hct116_df["infRV"] = infRV_Hct116
infRV_Hct116_df["mean"] = row_means
infRV_Hct116_df["variance"] = row_variances
infRV_Hct116_df = pd.DataFrame(infRV_Hct116_df)
infRV_Hct116_df
```

```python
# Plot histogram using Pandas
infRV_Hct116_df['infRV'].hist(bins=30, edgecolor='black')

# Add labels and title
plt.xlabel('infRV')
plt.ylabel('Frequency')
plt.title('Histogram of infRV')

# Show plot
plt.show()
```

```python
plt.boxplot(infRV_Hct116_df['infRV'], vert=False)

# Add labels and title
plt.xlabel('Values')
plt.title('Boxplot of infRV')

# Show plot
plt.show()
```

```python
stats = infRV_Hct116_df['infRV'].describe()
stats
```

```python
infRV_Hct116_df.to_csv('/mnt/scratch2/zahra/exp_quant/Hct116_quant/hct116_ShR/Hct116_31/aux_info/bootstrap/txp_infRV.csv', index=False, sep="\t")
```
